### Задача 2 (2 балла)

Посчитать 1! + 2! + 3! + 4! + ... + n! для заданного n. Решить только одним циклом

```python
def function_1(n):
    #Реализуйте последовательность действий, которая будет преобразовать число n в result 
    result = 0
    factorial = 1
    for i in range(1,n+1):
        factorial *= i
        result += factorial
    return result

#Теперь придумайте набор тестов, подтверждающих корректность вашего алгоритма

#Например
#assert function_1(1) == 1
#assert function_1(3) == 9 
```

```python
assert function_1(1) == 1
assert function_1(3) == 9
assert function_1(4) == 33
```

### Задача 3 (2 балла)

Найти среднее арифметическое отрицательных элементов в списке

```python
#list_of_numbers - список чисел
#mean - среднее арифметическое отрицательных элементов
def function_2(list_of_numbers):
    #тут реализуйте вычесление mean по list_of_numbers
    sum = 0
    mean = 0
    j = 0
    for a in list_of_numbers:
        if a < 0:
            j += 1
            sum += a
            mean = sum/j
    return mean

#Теперь придумайте набор тестов, подтверждающих корректность вашего алгоритма

#Например
#assert function_1([]) == 0
#assert function_1([1, -1, -3, 3]) == -2
# т.д. например, что будет, если список не содержит отрицательных элементов
```

```python
assert function_2([1, 2, 3, 4, -5]) == -5
assert function_2([1, -2, 3, 4, -5]) == -3.5
assert function_2([1, 2, 3, 4, 4]) == 0
assert function_2([-5]) == -5
assert function_2([]) == 0
```

### Задача 4 (2 балла)

Найти расстояние между точками с заданными координатами в n-мерном пространстве

```python
def function_3(point_1, point_2):
    #тут реализуйте вычисление расстояния distance между точками point_1 и point_2
    import math
    distance = 0
    for x, y in zip(point_1, point_2):
        distance += (x - y)**2
    distance = math.sqrt(distance)
    return distance

#Теперь придумайте набор тестов, подтверждающих корректность вашего алгоритма

#Например
#assert function_2((1, 0), (2, 0)) == 1
```

```python
assert function_3((1, 0, 1, 0), (1, 0, 0, 0)) == 1
assert function_3((1, 0), (2, 0)) == 1
assert function_3((1, 0, 0), (1, 0, 0)) == 0
```

### Задача 5 (3 балла)

Дан список-массив, заполненный случайным образом нулями и единицами. Найти самую длинную непрерывную последовательность единиц и определить индексы первого и последнего элементов в ней

```python
def function_sequence(list):
    begin_i = -1
    end_i = -1
    begin_index = -1
    end_index = -1
    len_ones_sequence = 0
    for j, a in enumerate(list):
        if (begin_i < 0):
            if (a == 1):
                begin_i = j
                end_i = j
        else:
            if (a == 1):
                end_i += 1
            else: 
                if (end_i - begin_i + 1 > len_ones_sequence):
                    end_index = end_i
                    begin_index = begin_i
                    len_ones_sequence = end_index - begin_index + 1          
                begin_i = -1
                end_i = -1
    if (begin_i >= 0):
        if (end_i - begin_i + 1 > len_ones_sequence):
            end_index = end_i
            begin_index = begin_i
            len_ones_sequence = end_index - begin_index + 1
    return len_ones_sequence, begin_index, end_index            
```

```python
assert function_sequence([0, 1, 1, 0, 1, 1, 1, 1, 0, 0, 1]) == (4, 4, 7)
assert function_sequence([0, 0, 0, 1]) == (1, 3, 3)
assert function_sequence([0, 0, 0, 0]) == (0, -1, -1)
assert function_sequence([1, 1, 1, 1]) == (4, 0, 3)
assert function_sequence([0, 1, 1, 0, 0, 1, 1, 0]) == (2, 1, 2)
```

Найти суммы элементов столбцов матрицы и отсортировать столбцы матрицы по возрастанию найденных сумм

```python
def function_5(matrix):
    #место для вашей реализации
    from copy import copy
    s = [0 for i in range(len(matrix[0]))] 
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            s[j] += matrix[i][j]
    ind = sorted(enumerate(s), key = lambda column:column[1])
    indexes, sums = zip(*ind)
    matrix_2 = [[0]*len(matrix[0]) for i in range(len(matrix))]
    for i in range(len(matrix)):
        for k, l in enumerate(indexes):
            matrix_2[i][k] = matrix[i][l]
    return matrix_2

#Тесты
#assert function_5([[5, 0, 1], [1, 1, 1], [2, 2, 0]]) == [[1, 1, 1], [2, 2, 0], [5, 0, 1]]
```

### Задача 7 (3 балла)

На вход подаётся строка S и буква w. Нужно развернуть в этой строке подстроку, заключенную между первой и последней буквой w, и вернуть получившуюся строку. Если данной буквы в строке нет, то вернуть -1. Если строка содержит только одну данную букву, то вернуть строку без изменений

```python
def function_6(S, w):
    #место для вашей реализации
    new_substring = ''
    ind_l = S.find(w)
    ind_r = S.rfind(w)
    if (ind_l == ind_r):
        if (ind_l < 0):
            return -1
        else:
            return S
    for i in range(ind_l + 1,ind_r):
        new_substring += S[-i+1]
    new_S = S[:ind_l+1] + new_substring + S[ind_r:]
    return new_S

#Тесты
#assert function_6('abcdefgefg', 'e') == 'abcdegfefg'
```

```python
assert function_6('abcdefgefg', 'e') == 'abcdegfefg'
assert function_6('abcdefgfg', 'e') == 'abcdefgfg'
assert function_6('abcdfgfg', 'e') == -1
```

### Задача 8 (4 балла + 2 балла за оформления в виде функции, принимающей текст на русском языке)

Дан текст. Посчитать, сколько раз встречается каждое словои и вывести слова в порядке убывания количествА повторений слова (при совпадении частоты встречаемости -- вывести в лексикографическом порядке)

```python
#with open('problem_8_text.txt', encoding="utf-8") as fl:
#    text = fl.read()
def frequency_dictionary(text):       
    text_fin = ''
    text_lower = text.lower().replace('\n\n',' ')
    for i in range(len(text_lower)):
        if (text_lower[i].islower()) or (text_lower[i] == ' ') or (text_lower[i] == '-'):
            text_fin += text_lower[i]
    words = text_fin.split(' ')
    words = [word for word in words if word != '']
    dictionary = {}
    for word in words:
        dictionary.setdefault(word, 0)
        dictionary[word] += 1        
    dict = sorted(dictionary.items(), key = lambda item:item[1], reverse = True)
    word, number = zip(*dict)
    word_numbers = sorted(list(set(number)), reverse = True)
    words_equal_number = []
    result = []
    for i in word_numbers:
        words_equal_number.clear()
        for w, n in dict:
            if n == i:
                words_equal_number.append(w)       
        words_equal_number.sort()           
        result.extend(words_equal_number)
    return(result)
             
        
#Теперь в text лежит сказка о Бабе-Яге =)

#Теперь обработайте текст и сделайте красивенький вывод
```
